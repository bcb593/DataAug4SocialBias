```python
import re
#less than 10 for 10k
#less than 11 for 50k

def filter_sentences(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    # Split the content into sentences using regular expressions to handle end of sentences more reliably
    sentences = re.split(r'(?<=[.!?])\s+', text)

    # Filter sentences based on word count and presence of alphabetic characters
    # This regex pattern checks for the presence of at least one alphabetic character
    valid_sentence_pattern = re.compile(r'[a-zA-Z]')

    shorter_sentences = [
        s.strip() for s in sentences 
        if 2 < len(s.split()) < 11
        and valid_sentence_pattern.search(s)  # Checks for at least one alphabetic character
    ]
    longer_sentences = [
        s.strip() for s in sentences 
        if len(s.split()) > 36 
        and valid_sentence_pattern.search(s)  # Checks for at least one alphabetic character
    ]

    return shorter_sentences, longer_sentences
```

```python
def check_len_range(sentences):
    # Calculate lengths in terms of words
    lengths = [len(sentence.split()) for sentence in sentences]

    min_length = min(lengths)
    max_length = max(lengths)

    print(f"The shortest sentence has {min_length} words.")
    print(f"The longest sentence has {max_length} words.")
```
